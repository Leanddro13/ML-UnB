# Modulo 03 - Regressão: Mínimos quadrados (least squares)
Queremos analisar como o tempo de estudo 
influencia na nota dos alunos. Para isso, vamos 
simular um conjunto de dados onde o tempo 
de estudo varia entre 0 a 10 horas e a nota 
varia de 0 a 10. 
Nossa hipótese é que quanto mais um aluno 
estuda, melhor sua nota. Vamos utilizar o 
método dos mínimos quadrados para 
encontrar a relação linear entre tempo de 
estudo e nota.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Passo 1: Gerar dados Sintéticos
np.random.seed(42) # Semente para reprodutibilidade
tempoDeEstudo = 10 * np.random.rand(100, 1) # tempo de estudo entre 0 e 10 horas
notaBase = 5 # Nota base assumida sem estudo
notaPorHora = 0.5 # Ganho de nota por hora de estudo

notas = notaBase + notaPorHora * tempoDeEstudo + np.random.randn(100, 1)
notas 
```
